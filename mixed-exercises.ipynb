**Write a code to find if a string is palindrome**

```python
test_string = "A man, a plan, a canal, Panama"
```

```python
def is_palindrome(s):
    s = "".join(filter(str.isalnum, s)).lower()
    return s == s[::-1]
result = is_palindrome(test_string)
```

```python
print(f"is the string {test_string} a palindrome? {result}")
```

**Write a code to find if words in a list are palindrome**

```python
list_of_strings = ["radar", "hello", "level", "world", "deified"]
```

```python
def find_palindrome(string):
    def is_palindrome(s):
        s = "".join(filter(str.isalnum, s)).lower()
        return s == s[::-1]
    return [s for s in string if is_palindrome(s)]
```

```python
palindromes = find_palindrome(list_of_strings)
print(palindromes)
```

**Write a function with iterative approach to find a factorial of a number**

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif  n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n+1):
            result *= i
        return result
```

```python
factorial(5)
```

**Write a function with recursive approach to find a factorial of a number**

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

factorial(5)
```

**Write a function with reduce/lambda approach to find a factorial of a number**

```python
from functools import reduce

def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        return reduce(lambda x, y: x * y, range(2, n +1))
    
factorial(5)
```
